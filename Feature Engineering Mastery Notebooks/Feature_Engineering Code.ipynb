# Feature Engineering vs Feature Selection vs Feature Extraction

```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.feature_selection import SelectKBest, f_regression

df = pd.DataFrame({
    "x1": [1, 2, 3, 4],
    "x2": [2, 4, 6, 8],
    "y": [10, 20, 30, 40]
})

df["ratio"] = df["x2"] / df["x1"]

poly = PolynomialFeatures(degree=2)
poly_features = poly.fit_transform(df[["x1", "x2"]])

selector = SelectKBest(score_func=f_regression, k=1)
selected = selector.fit_transform(df[["x1", "x2"]], df["y"])

poly_features, selected
```

# Mean and Median Imputation

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Age": [20, 25, np.nan, 40, 45],
    "Salary": [30000, 45000, 50000, np.nan, 70000]
})

df["Age"] = df["Age"].fillna(df["Age"].median())
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())

df
```

# Forward Fill and Backward Fill

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Temperature": [30, np.nan, np.nan, 35, 36]
})

forward_filled = df.fillna(method="ffill")
backward_filled = df.fillna(method="bfill")

forward_filled, backward_filled
```

# KNN Imputation

```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

df = pd.DataFrame({
    "Age": [25, 30, np.nan, 40],
    "Salary": [50000, 60000, 65000, np.nan]
})

imputer = KNNImputer(n_neighbors=2)
imputed_data = imputer.fit_transform(df)

pd.DataFrame(imputed_data, columns=df.columns)
```

# Regression-Based Imputation

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

df = pd.DataFrame({
    "Experience": [1, 2, 3, 4, 5],
    "Salary": [30000, 40000, np.nan, 60000, np.nan]
})

train = df.dropna()
test = df[df["Salary"].isna()]

model = LinearRegression()
model.fit(train[["Experience"]], train["Salary"])

df.loc[df["Salary"].isna(), "Salary"] = model.predict(test[["Experience"]])

df
```

# Z-Score Method

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Marks": [5, 6, 6, 6, 1999999999999998]
})

mean_value = df["Marks"].mean()
std_value = df["Marks"].std()

df["Z_Score"] = (df["Marks"] - mean_value) / std_value
print(df)
outliers = df[abs(df["Z_Score"]) > 1.5]

outliers
```

# IQR Method

```python
import pandas as pd

df = pd.DataFrame({
    "Salary": [10, 30000, 32000, 35000, 37000, 150000]
})

Q1 = df["Salary"].quantile(0.25)
Q3 = df["Salary"].quantile(0.75)
IQR = Q3 - Q1

outliers = df[
    (df["Salary"] < Q1 - 1.5 * IQR) | (df["Salary"] > Q3 + 1.5 * IQR)]

outliers
```

# Winsorization (Capping)

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Income": [30000, 32000, 35000, 37000, 150000]
})

lower = df["Income"].quantile(0.05)
upper = df["Income"].quantile(0.95)

df["Income_Capped"] = np.clip(df["Income"], lower, upper)

df
```

# Remove vs Keep vs Cap Decision

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Purchase_Amount": [500, 700, 800, 1200, 50000]
})

threshold = df["Purchase_Amount"].quantile(0.99)
print(threshold)

filtered = df[df["Purchase_Amount"] <= threshold]
capped = df.copy()
capped["Purchase_Amount"] = np.minimum(capped["Purchase_Amount"], threshold)

filtered, capped
```

# Variance Threshold

```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold

df = pd.DataFrame({
    "A": [1, 1, 1, 1],
    "B": [10, 12, 11, 13],
    "C": [5, 6, 7, 8]
})

selector = VarianceThreshold(threshold=0.0)
selected = selector.fit_transform(df)

selected
```

# Correlation-Based Feature Selection

```python
import pandas as pd

df = pd.DataFrame({
    "X1": [1, 2, 3, 4],
    "X2": [2, 4, 6, 8],
    "X3": [21, 69, 77, 51],
    "y": [10, 20, 30, 40]
})

correlation = df.corr()
correlation
```

# SelectKBest

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, f_regression

df = pd.DataFrame({
    "X1": [1, 2, 3, 4],
    "X2": [2, 1, 2, 1],
    "X3": [5, 6, 7, 8],
    "y": [10, 20, 30, 40]
})

X = df.drop("y", axis=1)
y = df["y"]

selector = SelectKBest(score_func=f_regression, k=2)
selected = selector.fit_transform(X, y)

selected
```

# Recursive Feature Elimination

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE

df = pd.DataFrame({
    "A": [1, 2, 3, 4],
    "B": [2, 3, 4, 5],
    "C": [5, 6, 7, 8],
    "y": [10, 20, 30, 40]
})

X = df.drop("y", axis=1)
y = df["y"]

model = LinearRegression()
rfe = RFE(model, n_features_to_select=2)
selected = rfe.fit_transform(X, y)

selected
```

# L1 Regularization (Lasso)

```python
import pandas as pd
from sklearn.linear_model import Lasso

df = pd.DataFrame({
    "X1": [1, 2, 3, 4],
    "X2": [2, 4, 6, 8],
    "X3": [1, 0, 1, 0],
    "y": [10, 20, 30, 40]
})

X = df.drop("y", axis=1)
y = df["y"]

model = Lasso(alpha=0.1)
model.fit(X, y)

model.coef_
```

# Tree-Based Feature Importance

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

X = pd.DataFrame({
    "Age": [22, 30, 45, 50],
    "Income": [30000, 50000, 90000, 120000],
    "Score": [650, 700, 720, 680]
})

y = [0, 1, 1, 1]

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

model.feature_importances_
```
