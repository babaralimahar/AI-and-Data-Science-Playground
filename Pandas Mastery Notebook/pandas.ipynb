## Handling Missing Values
Real-world datasets often have missing data. Pandas offers:
- `isna()` / `isnull()` ‚Üí detect missing values (returns True/False)
- `dropna()` ‚Üí remove missing values (rows/columns)
- `fillna()` ‚Üí replace missing values with a given value or method (e.g., forward fill)

Always inspect how much data is missing before deciding to drop or fill.

````python
data = {
    "Name": [
        "Babar", "Ali", "Sohail", "Ghulam", "Akram",
        "Ahmed", "Usman", "Hassan", "Bilal", "Adeel"
    ],
    "Age": [
        22, 25, np.nan, 30, 28,
        np.nan, 24, 27, 26, np.nan
    ],
    "City": [
        "Karachi", "Lahore", "Karachi", np.nan, "Islamabad",
        "Lahore", np.nan, "Peshawar", "Quetta", "Karachi"
    ],
    "Score": [
        85, np.nan, 78, 92, np.nan,
        88, 75, np.nan, 90, 82
    ]
}

df2 = pd.DataFrame(data)
df2
````

````python
df2.isnull()
````

````python
df2.isnull().sum()
````

````python
df2.dropna()
````

````python
df2['Age'] = df2['Age'].fillna(df2['Age'].mean())
df2['City'] = df2['City'].fillna(df2['City'].mode())
df2['Score'] = df2['Score'].fillna(df2['Score'].mode())
df2
````

##  Boolean Indexing & Conditional Selection
- Use comparison operators to build conditions (e.g., `df['col'] > 50`).
- Pass the condition inside `df[...]` to get matching rows.
- Combine multiple conditions with `&` (AND), `|` (OR), and `~` (NOT).
- Parentheses are required when combining conditions.

````python
data = {
    "Name": [
        "Babar", "Ali", "Sohail", "Ghulam", "Akram",
        "Ali", "Hassan", "Bilal", "Usman", "Babar"
    ],
    "Age": [
        22, 25, 28, 30, 28,
        25, 27, np.nan, 24, 22
    ],
    "Score": [
        85, 88, 78, 92, 88,
        88, 75, 90, np.nan, 85
    ]
}

df3 = pd.DataFrame(data)
df3
````

````python
df3[df3['Age'] < 25]
````

````python
df3[(df3["Age"] > 25) & (df3["Score"] > 35)]
````

````python
df3[(df3["Score"] < 25 ) | (df3["Score"] > 85)]
````

````python
# Not `~`
df3[~(df3["Score"] > 85)]
````

## Filtering Rows with Multiple Conditions

##  Concept of Split‚ÄìApply‚ÄìCombine
- **Split** the data into groups based on some criteria.
- **Apply** a function (e.g., sum, mean) to each group.
- **Combine** the results into a new DataFrame or Series.

This is the core idea behind `groupby`.

````python
e = {
    "Employee": [
        "Babar", "Ali", "Sohail", "Ghulam", "Akram",
        "Ali", "Hassan", "Bilal", "Usman", "Babar"
    ],
    "Department": [
        "IT", "HR", "Finance", "IT", "Finance",
        "HR", "IT", "Sales", "Sales", "Finance"
    ],
    "Salary": [
        70000, 60000, 65000, 72000, 68000,
        60000, 71000, np.nan, 55000, np.nan
    ]
}

dat = pd.DataFrame(e)
dat
````

````python
dat['Department'].value_counts()
````

````python
dat[dat['Department'] == "IT"]["Salary"]
````

````python
dat[dat['Department'] == "IT"]["Salary"].mean()
````

````python
dat.groupby("Department")["Salary"].mean() #Shortcut Way
````

````python
dat.groupby('Department')["Employee"].sum()
````

##  Using `groupby()` with Aggregation Functions
- Use `groupby('col').mean()` or other functions like sum, count, min, max.
- You can also use `agg()` to pass multiple functions at once.

````python
dat.groupby("Department")["Salary"].agg(["mean","max","min"])
````

## Using `concat()` for Stacking Data
- `pd.concat()` combines DataFrames vertically (rows) or horizontally (columns).
- Use `axis=0` for vertical stacking (default), `axis=1` for horizontal.
- `ignore_index=True` can reset the index after concatenation.

````python
data = {
    "ProductID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "ProductName": [
        "Laptop", "Mobile", "Tablet", "Headphones", "Monitor",
        "Keyboard", "Mouse", "Speaker", "Charger", "Smartwatch"
    ],
    "Price": [1200, 800, 450, 150, 300, 100, 50, 150, 50, 300]
}

df5 = pd.DataFrame(data)


data2 = {
    "ProductID": [201, 202, 203, 204, 205, 206, 207, 208, 209, 210],
    "ProductName": [
        "Printer", "Router", "SSD", "Hard Drive", "Power Bank",
        "USB Cable", "Webcam", "Microphone", "Projector", "AirPods"
    ],
    "Price": [250, 120, 180, 200, 150, 25, 90, 110, 500, 180]
}

df6 = pd.DataFrame(data2)
````

````python
df5, df6
````

````python
pd.concat([df5,df6], ignore_index= True)
````

##  Using `merge()` for SQL-Style Joins

- Specify `on` or `left_on`/`right_on` to choose key columns.
- Join types: `inner` (default), `left`, `right`, `outer`.

````python
products = pd.concat([df5,df6], ignore_index= True)
products
````

````python
sales_data =pd.DataFrame( {
    "ProductID": [101, 102, 103, 104, 105, 101, 102, 108, 109, 110], # 106 Doesn't exist in Products
    "UnitsSold": [5, 12, 7, 20, 9, 8, 15, 6, 10, 4]
})

sales_df = pd.DataFrame(sales_data)
sales_df
````

````python
pd.merge(products, sales_data, how= "right")
````

````python
pd.merge(products, sales_data, how = "outer")
````

##  Using `join()` with Index Alignment
- `join()` joins DataFrames on their index or on a key column.

##  Applying Custom Functions with `apply()` / `map()`
- `apply()` lets you apply a custom function across rows or columns.
- `map()` is for applying functions element-wise to a Series.
- Both are handy for transformations, calculations, and cleaning data.

````python
student = {
    "Name": ["Ali", "Ayesha", "Babar", "Zainab", "Usman", "Hina", "Ahmed", "Sara", "Hamza", "Noor"],
    "Math": [85, 78, 92, 66, 88, 74, 90, 81, 69, 95],
    "Science": [80, 82, 89, 70, 85, 76, 92, 79, 72, 94]
}

df = pd.DataFrame(student)
df
````

````python
df["Total Marks"] = df[["Math","Science"]].apply(lambda row: row.sum(),axis=1)
df["Total Marks"]
````

````python
col_mean = df[["Math","Science"]].apply(lambda col: col.mean())
col_mean
````

````python
def to_grade(s):
    if s>=90:
        return "A"
    elif s>=70:
        return "B"
    elif s>=40:
        return "C"
    else:
        return "F"
    
df["Math Grade"] = df["Math"].apply(to_grade)
````

````python
df
````

````python
df["Name"]= df["Name"].map(str.lower)
df
````

````python
df["Bonus Math"] = df["Math"].map(lambda x: x+5)
df
````
